```python
# Cargamos las librerias necesarias
import pandas as pd
import numpy as np
from pathlib import Path
```

## 1. Carga Datos

```python
## Directorio donde está el notebook
BASE_DIR = Path.cwd()
# Cargar datos. Se cargan los inputs limpios, pero sin feature engineering
df_total = pd.read_csv(f"{BASE_DIR}/SET_INPUTS_BASE_2.csv", sep=',', encoding='latin-1')
```

###  Para determinar el mejor modelo, hacemos el tratamiento de variables fuera del Pipeline final, generamos VARIABLES nuevas que pueden ser de utilidad (Feature Engineering) 
# INPUTS_Trat_2. Son la entrada a la optimización del modelo

```python
# 2. Creación de nuevas variables
# Nuevas variables
df_total ['Variac_Temp'] = abs(df_total ['Temp_media_C_Nat'] - 16)
df_total ['Gen_E_Relat'] = df_total ['Gen_Eolica_GWh'] / df_total ['Demanda_GWh']
df_total ['Gen_S_Relat'] = df_total ['Gen_Solar_GWh'] / df_total ['Demanda_GWh']

# Índice del gas (base 2021 = 100)
# Precio del gas en el año base (tomamos el de 2021 al igual que en el IPC)
precio_base = df_total.loc[df_total['Year'] == 2021, 'Price_Gas_EUR_MWh'].mean()
print(precio_base)
df_total['IndiceGas'] = (df_total['Price_Gas_EUR_MWh'] / precio_base) * 100


# Variables temporales
df_total = df_total.sort_values(["Year", "Month", "Day"]).reset_index(drop=True)
df_total["time_idx"]=np.arange(len(df_total))
df_total["sin_doy"] = np.sin(2 * np.pi * df_total["DayOfYear"] / 365.25)
df_total["cos_doy"] = np.cos(2 * np.pi * df_total["DayOfYear"] / 365.25)


df_total.to_csv(f"{BASE_DIR}/SET_INPUTS_TRAT_2.csv", index=False)
```
