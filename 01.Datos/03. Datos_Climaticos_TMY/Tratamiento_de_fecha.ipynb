### Ajuste temporal del dataset TMY al año 2026

En este bloque de código se carga el fichero `TMY_daily_agg.csv`, que contiene datos climáticos diarios agregados (TMY). La columna de fecha (`date_utc`) se convierte a formato `datetime` para permitir su manipulación temporal y, posteriormente, se reemplaza el año original de todas las observaciones por **2026**, manteniendo inalterados el mes y el día.  

Este procedimiento permite reutilizar un **año meteorológico típico** como escenario climático representativo para 2026, sin modificar la estructura estacional de los datos. A continuación, el conjunto de datos se ordena por latitud y fecha para asegurar una correcta coherencia temporal y espacial. Finalmente, el DataFrame resultante se exporta al fichero `TMY_daily_agg_2026.csv`, que será utilizado en las fases posteriores del análisis.

```python
import pandas as pd
import numpy as np

df = pd.read_csv("TMY_daily_agg.csv")

# Convertir la columna de fecha a formato datetime
df["date_utc"] = pd.to_datetime(df["date_utc"], format="%Y-%m-%d")

# Cambiar el año de las fechas a 2026
df["date_utc"] = df["date_utc"].apply(lambda x: x.replace(year=2026))

# Ordenar el DataFrame por latitud y fecha 
df = df.sort_values(["lat", "date_utc"], ascending=[True, True]).reset_index(drop=True)


df.to_csv("TMY_daily_agg_2026.csv", index=False)
print("Archivo guardado: TMY_daily_agg_2026.csv")
```
