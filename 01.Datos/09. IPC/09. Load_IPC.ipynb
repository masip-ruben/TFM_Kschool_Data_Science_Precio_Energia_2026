```python
# Cargamos las librerias necesarias
import pandas as pd
from pathlib import Path
```

```python
# Directorio donde est√° el notebook
BASE_DIR = Path.cwd()
# Cargar datos
df_ipc = pd.read_excel(f"{BASE_DIR}/IPC Nacional 2002-2025 mes a mes.xls", skiprows=6)

df_ipc.drop(columns=['Unnamed: 0'], inplace=True)
```

## Crear una dataframe con formato daily

```python
df_ipc_daily=[]
for i in range(len(df_ipc)):
    year = df_ipc.iloc[i, 0]
    for month in range(1, 13):
        days_in_month = pd.Period(f'{year}-{month:02d}').days_in_month
        daily_value = df_ipc.iloc[i, month]
        for day in range(1, days_in_month + 1):
            df_ipc_daily.append({'Year': year, 'Month': month, 'Day': day, 'IPC Diario': daily_value})
df_ipc_daily = pd.DataFrame(df_ipc_daily)
df_ipc_daily.head()
# Guardar resultados
df_ipc_daily.to_csv(f"{BASE_DIR}/IPC_daily.csv", index=False)
```
